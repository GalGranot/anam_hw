## <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 1 - Matrix derivatives
---
####  Vector & Matrix Deriviatives
* $\nabla_x Ax = A^{T}$
* $\nabla_x x^{T} A x = (A + A^{T}) x$ 
* $\frac{\partial}{\partial A} \ln |A| = A^{-T}$
* $\frac{\partial}{\partial A} Tr[AB] = B^{T}$

Using the above, we will use the following:
1. $\nabla_{\mu} {\mu}^{T} \Sigma^{-1} x_i = \Sigma^{-1} x_i$ 
2. $\nabla_{\mu} {\mu}^{T} \Sigma^{-1} \mu = (\Sigma^{-1} + {\Sigma}^{-T}) \mu$
3. $\frac{\partial}{\partial \Sigma^{-1}} \ln |\Sigma^{-1}| = \Sigma^{T} = \Sigma$
4. $\frac{\partial}{\partial \Sigma^{-1}} Tr[\Sigma^{-1} \sum_{i=1}^n (\overline{x_i} - \overline{\mu}) (\overline{x_i} - \overline{\mu})^{T}] = \sum_{i=1}^n (\overline{x_i} - \overline{\mu}) (\overline{x_i} - \overline{\mu})^{T}$
---
####  Question time

Let $X_1,\ldots,X_N\sim \mathcal{N}(\mu,\Sigma)$ i.i.d.,$\ \mu\in \mathcal{R}^d,\ \Sigma\in \mathcal{R}^{d\times d}$ .

1. Find the MLE estimator $\hat{\mu}_{MLE}$. 
    * Reminder: finding the MLE estimator is done by comparing the gradient to zero.
    * Use the vector and matrix derivatives above. 
2. Find the MLE estimator $\hat{\Sigma}_{MLE}$
     * You should use **The Trace Trick** - $\sum_{i=1}^n (\overline{x_i} - \overline{\mu})^{T} \Sigma^{-1} (\overline{x_i} - \overline{\mu}) = \sum_{i=1}^n \textit{Trace}\big((\overline{x_i} - \overline{\mu})^{T} \Sigma^{-1} (\overline{x_i} - \overline{\mu})\big) = \textit{Trace}\big(\Sigma^{-1} \sum_{i=1}^n (\overline{x_i} - \overline{\mu}) (\overline{x_i} - \overline{\mu})^{T} \big)$ .
     * It may be easier to use $\frac{\partial}{\partial \Sigma^{-1}}$ .

#### <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 7.3 - Robustness to Noise & Outliers
---
- Explain the results. In your answer, describe the effect of adding noise and outliers, which algorithm is more affected and why?

PCA on noisy data:
We can see PCA managed to separate the data pretty well, the class labels are more spread out and their clusters intersect more than the original no-noise pca, but still works.

PCA aims to find the directions (principal components) that maximize the variance in the data. Adding noise increases the overall variance, which can slightly distort the principal components.



PCA on skewed data:
All the random samples are at a cluster far away from the other, real classes, but the clusters for the real classes are less seperated than original and even noisy (which means that even though pca managed to figure out the skewed data, it still hurt the dimension reduction for the other classes).

PCA is sensitive to outliers because it relies on variance. The addition of skewed data (random samples) increases the overall variance significantly, which can distort the principal components more than noise.


TSNE on noisy data:
Making sure to tune the perplexity (ending at 30), we can hardly see any difference from the original no noise dimension reduction, TSNE is pretty robust to noise, unlike PCA.

t-SNE is a non-linear method designed to preserve local structures in high-dimensional data by converting pairwise distances to probabilities and minimizing the difference between these probabilities in lower dimensions. Noise affects these probabilities less significantly, allowing t-SNE to maintain the structure and clustering of the data even with added noise.


TSNE on skewed data:
Managed to diffrentiate the skewed data samples very well and still kept the separation of the other classes. TSNE's emphasis on local neighborhoods allows it to isolate outliers effectively while maintaining the separation of real class clusters, making it more robust to outliers than PCA.
